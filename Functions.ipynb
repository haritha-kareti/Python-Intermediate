```python
def sharuk(n):
  s = 0
  for j in range(1,n):
    if n % j == 0:
      s = s + j

  if s == n:
    print(f"{n} : perfect")
  else:
    print(f"{n} : Not Perfect")

sharuk(6) # calling
```

difference between local varibales and global varibales

```python
r = 100 # global varibales

def kiran():
  print(f'I am from Inside the function : {r}')

kiran() # calling

print(f'I am from Outside the function : {r}')
```

```python
def kiran():
  x = 500 # local varbale
  print(f'I am from Inside the function : {x}')

kiran() # calling

print(f'I am from Outside the function : {x}')
```

```python
def kiran():
  global x
  x = 500 # local varbale to global
  print(f'I am from Inside the function : {x}')

kiran() # calling

print(f'I am from Outside the function : {x}')
```

```python
def fun(n):
  for i in range(1,7):
    if i % 2 == 0:
      print(f'Even : {i}')

  c = 0
  for j in range(1,n+1):
    if n % j == 0:
      c = c + 1

  if c == 2:
    print(f'Prime')
  else:
    print(f'Not prime')

fun(7) # calling
```

```python
fun(8)
```

```python
def sharuk(a,b):
  c = a + b
  print(c)


sharuk(10,20) # calling
```

```python
def sharuk(a,b):
  c = a + b
  return c # return is used to return back the result to the place where you call
             # The function


sol = sharuk(10,20) # calling
print(sol)
```

```python
def sharuk(a,b):
  c = []
  for i in range(a,b):
    if i % 2 == 0:
      c.append(i)

  return c


sol = sharuk(10,20) # calling function
print(sol)
```

modularization using functions

```python
def sharuk(a,b):
  print(a+b)



sharuk(10,20) # calling the function
```

```python
def sharuk(a = None, b = None):
  print(a+b)



sharuk(10,20) # calling the function
```

```python
def sharuk(a = None, b = None):
  print(a)
  print(b)



sharuk(10) # calling the function
```

```python
def sharuk(a = None, b = None):
  print(a)
  print(b)



sharuk(b = 10) # calling the function
```

```python
def kiran(f,g):
  print(f)
  print(g)



kiran([10,20,30],[1,2,3])
```

```python
def sharuk(*a):
  for i in a:
   for j in i:
    if j%2 != 0:
      print(j)

sharuk([10,11,12],[91,11,22],[1,2,3]) # calling the function
```

```python
def sharuk(a):
  print(a)
  for i in a.values():
    print(i)


fruits = {'apple':100,'Banana':200,'Carrot':300}
sharuk(fruits)
```
