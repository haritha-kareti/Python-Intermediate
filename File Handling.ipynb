**File Handling**

`File Handling concept is used to work with different Files (ex: pdf | pptx | txt | word_doc | csv )`

`In File Handling concept -> we have 3 Modes `

- `Mode w -> write Mode `
- `Mode a -> add Mode (update)`
- `Mode r -> read Mode `

**Mode w**

- `Rule 1: If the File is Not there Mode w will create the file and we can add the data into the file `

- `Rule 2: If the File is Available and File is Empty Mode w can write the data into the file `

- `Rule 3: If the File is Available and There is data in the File but if we pass New data Old data will override `

```python
a = open('sharuk.txt' , 'w') # write Mode

a.write('Data science')

a.close()
```

- `dump -> sending the dict into a json file `
- `load -> read the dict from a Json file `
- `dumps -> convert dict to str `
- `loads -> convert str back to dict `

```python
# JSON = Java Script Object notation
```

```python
import json
```

```python
fruits = {'apple':100,'Banana':200,'Carrot':300}

with open('kumar.json','w') as f:
  json.dump(fruits,f)
```

```python
with open('kumar.json','r') as f:
  result = json.load(f)

print(result)
```

```python
fruits = {'apple':100,'Banana':200,'Carrot':300} # dict
print(fruits)
print(type(fruits))


res = json.dumps(fruits)
print(res)
print(type(res))
```

```python
fruits = {'apple':100,'Banana':200,'Carrot':300} # dict
print(fruits)
print(type(fruits))


res = json.dumps(fruits) # dict to str
print(res)
print(type(res))


final = json.loads(res) # str to dict
print(final)
print(type(final))
```
