**Lambda Function**


- `Lambda function is known as Ananymous Function`
- `Ananymous Function Means -> A function without a function name `
- `syntax of lambda function : lambda arguments : expressions `


```python
a = 10
b = 20
print(a + b)
```

```python
def kiran(a,b):
  print(a+b)

kiran(10,20) # calling
```

```python
f = lambda a,b : a + b

print(f(10,20))
```

```python
n = 5

for i in range(1,11):
  print(f'{n} * {i} = {n*i}')
```

```python
def table(n):
  for i in range(1,11):
    print(f'{n} * {i} = {n*i}')


table(5)
```

```python
table(2)
```

```python
n = 2

f = lambda x,y : x*y

for i in range(1,11):
  print(f'{n} * {i} = {f(n,i)}')
```

Map | Filter | reduce

```python
# map syntax = map(function_name,arguments)

def fun(b):
  if b%2 == 0:
    return b



sol = map(fun,[10,11,12,13,14,15])
print(list(sol))
```

```python
def fun(b):
  if b%2 == 0:
    return b


sol = filter(fun,[10,11,12,13,14,15])
print(list(sol))
```

```python
sol = map(lambda x : x % 2 == 0 ,[10,11,12,13,14,15])
print(list(sol))
```

```python
sol = filter(lambda x : x % 2 == 0 ,[10,11,12,13,14,15])
print(list(sol))
```

```python
import functools

f = [1,4,7,9,3,6]

sol = functools.reduce(lambda x,y : x+y , [1,4,7,9,3,6])
print(sol)
```

**Map with Filter**

```python
sol = map(lambda x:x+20 , filter(lambda x:x%2==0,[10,11,21,22,26,36,35]))
print(list(sol))
```

**Filter with Map**

```python
sol = filter(lambda x:x%2==0 , map(lambda x:x+20,[10,11,21,22,26,36,35]))
print(list(sol))
```

**reduce with map and filter**

```python
sol = functools.reduce(lambda x,y:x+y,map(lambda x:x+20 , filter(lambda x:x%2==0,[10,11,21,22,26,36,35])))
print(sol)
```
